```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dt = pd.read_csv("prem.csv")
dt
```

```python
x = dt.Team.head(5)
y = dt.Points.head(5)
z = dt.GF.head(5)
plt.plot(x, y)
plt.plot(x, z)
plt.title("Premier League 21/22 szn top 5 teams")
plt.xlabel("Teams")
plt.ylabel("Points and Goals scored")
plt.legend(["Points", "Goals"])
plt.show()
```
