```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
```

```python
ok = urlopen("https://oneokrock.fandom.com/wiki/Song")
bsobj = soup(ok.read())
```

```python
bsobj.findAll('a')
```

```python
for link in bsobj.findAll('a'):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
for link in bsobj.find('div',{'class':'mw-parser-output'}).findAll('a',href = re.compile('^(/wiki/)((?!:).)*$')):
    if 'href' in link.attrs:#here i only want to get songs that linked to a website
        print(link.attrs['href'])
```

```python
import pandas as pd
import json
```

```python
okr = []
for link in bsobj.find('div',{'class':'mw-parser-output'}).findAll('a',href = re.compile('^(/wiki/)((?!:).)*$')):
    if 'href' in link.attrs:
        okr.append(link.attrs['href'])
print(okr)
```

```python
df = pd.DataFrame(okr) #create dataframe
print(df)
```

```python
okr = df.to_dict()
type(okr)
okr
```

```python
with open('OneOkRock_Song_Link.json', 'w', encoding='utf-8') as file:#save in json file
    json.dump(okr, file, ensure_ascii=False) 
```
