```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('Marvel Movies.csv')
dt
```

```python
subdata = dt[['name', 'HAIR']]
subdata
```

```python
top = subdata.groupby(['name'])['HAIR'].apply(lambda x: ','.join(str(i) for i in x)).reset_index()
top
```

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = top['HAIR'].tolist()
vec = CountVectorizer()
X = vec.fit_transform(docs)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df
```

```python
coocc = df.T.dot(df)
coocc
```

```python
import numpy as np

filter_num = 0
getid = coocc.index
important = pd.DataFrame(columns = getid).T
important['relation'] = ""
subcoocc = coocc
dropid = []

for i in range(len(coocc)):
    result = coocc[coocc.iloc[i] > filter_num].index.tolist()    
    if(len(result) > 0):
        important['relation'].loc[getid[i]] = coocc[coocc.iloc[i] > filter_num].index.tolist()
    else:
        dropid.append(i)
        important = important.drop([getid[i]])
        subcoocc = subcoocc.drop([getid[i]])    
subcoocc
```

```python
subcoocc.drop(subcoocc.columns[dropid], axis=1, inplace=True)
subcoocc
```
