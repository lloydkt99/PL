```python
import pandas as pd

prem = pd.read_csv("prem.csv")
prem
```

```python
#2 sort the teams in alphabethical order
sort = prem.sort_values(["Team"])
sort
```

```python
#3 sort teams from most win to least
most = prem.sort_values(["Won"], ascending=False)
most
```

```python
#4 i want to know manchester united data 21/22 szn
#using location
man_utd = print(prem.iloc[5])
man_utd
```

```python
#5 other way looking for specific team data
#using teams name
spec_team = prem.loc[prem['Team'] == "Man Utd"]
spec_team
```

```python
#6 looking for 3 teams that drawn the most
most_draw = prem.sort_values(by='Drawn', ascending=False).iloc[0:3]
most_draw
```

```python
#7 looking for teams that have won more than 10 times and drawn less than 10 and lost less than 15
test = prem.loc[(prem['Won'] > 10) & (prem['Drawn'] < 10) & (prem['Lost'] < 15)]
test
```

```python
#8 total goal scored 21/22 szn(GF = Goal For)
total = prem['GF'].sum()
print(f"Total goal scored = {total}")
```

```python
#9 average goal scored per team in the 21/22 szn
avg = total / 20
print(f"Average goal scored in 21/22 szn is {avg}")
```

```python
#10 what is the goal per match in the 21/22 szn
#we know we have 20 teams that have played 38 games
tm = 38 * 20
goal = total / tm
print(f"Goal per match in the 21/22 premier league season is {goal}")
```
